# Task
Design and implement a Rule-Based Expert System in R for a Weather-Based Outfit Recommender. The system should use conditional rules to suggest clothing or gear based on user-provided weather conditions.

```python
def get_weather_input():
    print("ğŸŒ¦ï¸ Weather-Based Outfit Recommender ğŸŒ¤ï¸")
    print("\nPlease provide the following weather details:")

    try:
        temperature = float(input("ğŸŒ¡ï¸ Temperature (Â°C): "))
        condition = input("ğŸŒ¥ï¸ Condition (sunny, rainy, cloudy): ").strip().lower()
        time_of_day = input("ğŸ•’ Time of day (morning, afternoon, evening, night): ").strip().lower()
    except ValueError:
        print("\nâŒ Invalid input. Please enter a valid number for temperature.")
        return None

    return {
        "temperature": temperature,
        "condition": condition,
        "time_of_day": time_of_day
    }


def outfit_recommender(weather):
    temperature = weather["temperature"]
    condition = weather["condition"]
    time = weather["time_of_day"]

    recommendations = []

    # --- Temperature-Based Rules ---
    if temperature < 0:
        recommendations += ["Heavy winter coat", "Thermal wear", "Gloves and beanie"]
    elif 0 <= temperature < 10:
        recommendations += ["Warm coat", "Scarf and boots"]
    elif 10 <= temperature < 20:
        recommendations.append("Light jacket or sweater")
    elif 20 <= temperature < 30:
        recommendations.append("T-shirt and jeans or shorts")
    else:
        recommendations += ["Very light clothing", "Stay hydrated"]

    # --- Weather Condition Rules ---
    if condition == "rainy":
        recommendations += ["Carry an umbrella or wear a raincoat", "Waterproof shoes"]
    elif condition == "sunny":
        recommendations += ["Wear sunglasses", "Use sunscreen"]
        if time == "afternoon":
            recommendations.append("Cap or hat for sun protection")

    # --- Time of Day Rule ---
    if time in ["evening", "night"] and temperature < 15:
        recommendations.append("It might get colder â€” carry an extra layer")

    return recommendations


def main():
    weather = get_weather_input()
    if weather is None:
        return

    suggestions = outfit_recommender(weather)

    print("\nğŸ‘• Based on the weather, you should consider wearing:")
    print("-" * 45)
    for item in suggestions:
        print(f"âœ… {item}")
    print("-" * 45)
    print("âœ… Stay safe and dress appropriately! ğŸ˜Š")


if __name__ == "__main__":
    main()
```

If it's evening and chilly, it reminds the user to carry an extra layer.

The goal is to simulate expert advice using simple condition-based logic.
